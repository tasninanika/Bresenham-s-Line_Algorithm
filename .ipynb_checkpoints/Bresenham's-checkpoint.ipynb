```python
import matplotlib.pyplot as plt

def bresenham(x1, y1, x2, y2):
    x, y = x1, y1
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1
    err = dx - dy
    points = [(x, y)]

    while x != x2 or y != y2:  
        e2 = 2 * err
        if e2 > -dy:
            err -= dy
            x += sx
        if e2 < dx:
            err += dx
            y += sy
        points.append((x, y))  

    return points

# Example usage
x1, y1 = 2, 3
x2, y2 = 10, 8
points = bresenham(x1, y1, x2, y2)

# Extract x and y coordinates
x_vals, y_vals = zip(*points)

# Plot with connected lines
plt.plot(x_vals, y_vals, marker='o', color='purple', linestyle='-')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bresenham’s Line Algorithm")
plt.grid()
plt.show()
```
